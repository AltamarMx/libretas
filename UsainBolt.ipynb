```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
%matplotlib inline
```

### Sencillo programa en Python para graficar y calcular la energía utilizada por Usain Bolt en su mítica carrera de 100m y que se presenta en el webinar ¿Cuánta energía uso al ejercitarme? G. Barrios y J.A. del Río, Coordinadora C. Morales IER-UNAM

### Implementación de la fórmula para estimar las **Energía** (Joule) usando las ecuaciones presentadas en el artículo "On the performance of Usain Bolt in the 100m sprint" http://dx.doi.org/10.1088/0143-0807/34/5/1227

```python

#Define el intervalo de tiempo de 0 a 9.53
t = np.arange(0.01,9.53,0.01)
#Define constantes de la soluci'on
m = 86.
A = 110.
B = 12.2
k = 0.9
#calcula la posici'on como funci'on del tiempo
x = A/k *np.log( (A+B*np.exp(-k*t) )/(A+B)) + B/k*np.log((A*np.exp(k*t) +  B)/(A+B) )
#calcula la potencia como funci'on del tiempo
P = 1285.1489666307857*m*A*B*k*(A+B)*(((1.-np.exp(-k*t))*np.exp(-k*t)) )/(A+(B*np.exp(-k*t)))**3.
```

```python
#grafica posici'on y potencia
fig, ax  = plt.subplots(2,1,figsize=(10,8),sharex=True)

ax[0].plot(t,x)
ax[0].grid()
ax[0].set_ylabel("x [$m$]")


ax[1].plot(t,P)
ax[1].grid()
ax[1].set_ylabel("Potencia [$W$]")
ax[1].set_xlabel("tiempo [$s$]")
ax[1].set_xticks(np.arange(0,11,1)) 
```

```python
print("La energía utilizada por Usain Bolt es: ",round(integrate.simps(P,t)/4.188/1000.,3),"kcalorías")
```
